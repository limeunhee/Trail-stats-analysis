```python
import glob
from bs4 import BeautifulSoup
from multiprocessing import Pool
import numpy as np
import pandas as pd
```

```python
def try_except(prompt):
    """ 
    This function tries out prompt, and appends the result to lst

    Args:
        prompt (str): an expression to evaulate
        lst (List[Object]): a list where output to prompt is appened to

    Returns:
        lst

    """

    try:
        #print("in try")
        a = eval(prompt)
        return a
    except:
        #print("in except")
        return None
```

```python
# Defining class names
class_name = "styles-module__containerDescriptive___3aZqQ styles-module__trailCard___2oHiP"
class_name_2 = "xlate-none styles-module__link___12BPT"
class_name_4 = "styles-module__content___1eARw styles-module__content___3dWXB"
class_name_6 = "styles-module__info___1Mbn6"
class_name_8 = "xlate-none"
```

```python
def get_trail_urls(state_html):
    """ 
    This function takes in a filepath for a state html and 
    i) saves a file with urls of trails in that state from Alltrails.com 
    ii) saves a csv file containing duration and distance of trails from the state 

    Args:
        state_html (str): a filepath for a file that contains html of trails search result for a state
    Returns:
        None
    """

    state = state_html[37:].split(".")[0]
    print(state)

    soup = BeautifulSoup(open(state_html), "html.parser")

    each_urls = []
    durations = []
    distances = []

    div = soup.find_all("div", {"class": class_name})

    for container in div:

        # Get list of url to .txt file
        each_url = "https://www.alltrails.com" + (container.a["href"])
        each_url = each_url.replace("?ref=result-card", "")
        each_urls.append(each_url)

        # Save list of url to .txt file
        filename = "./" + state + ".txt"
        np.savetxt(filename, each_urls, fmt="%s", delimiter=",")

        div6 = container.find_all("div", {"class": class_name_6})

        # Get list of durations for each of 500 trail
        try:
            print("in try")
            duration = div6[1].find_all('span', {'class': 'xlate-none'})[1].text[5:]
        except:
            print("in except")
            duration = None
        durations.append(duration)
#         duration = "div6[1].find_all('span', {'class': 'xlate-none'})[1].text[5:]"
#         durations.append(try_except(duration))

        # Get list of distances for each of 500 trail
        try:
            print("in try")
            distance = div6[1].find_all('span', {'class': 'xlate-none'})[0].text[8:]
        except:
            print("in except")
            distance = None
        distances.append(distance)
#         distance = "div6[1].find_all('span', {'class': 'xlate-none'})[0].text[8:]"
#         distances.append(try_except(distance))

        filename = "./" + state + ".csv"
        #np.savetxt(filename, np.c_[distances, durations], fmt="%s", delimiter=",")
        pd.DataFrame({"distances": distances, "durations": durations}).to_csv(filename)
        
    return
```
